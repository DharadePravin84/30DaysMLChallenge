```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor ,plot_tree
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load Iris dataset
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
iris.data
```

```python
X = iris.data
y = iris.target
```

```python
iris.target_names
```

```python
# Use a specific feature for regression (petal width)
X_reg = X[:, 2]
```

```python
# Create a DataFrame for better visualization
iris_df = pd.DataFrame(data=np.c_[X_reg, y], columns=['Petal Width', 'Target'])
```

```python
iris_df.head()
```

```python
iris_df.shape
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_reg, y, test_size=0.2, random_state=42)
```

```python
# Reshape for single feature
X_train = X_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
```

```python
# Decision Tree Regression
regressor = DecisionTreeRegressor(random_state=42)
regressor.fit(X_train, y_train)
```

```python
# Predictions
y_pred = regressor.predict(X_test)
```

```python
# Evaluation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
# Plotting the actual data
sns.scatterplot(x=X_test.flatten(), y=y_test, color='blue', label='Actual', s=80)
```

```python
# Plotting the predictions
sns.scatterplot(x=X_test.flatten(), y=y_pred, color='red', label='Predicted', s=80)
```

```python
# Visualization
# Plotting the actual data
sns.scatterplot(x=X_test.flatten(), y=y_test, color='blue', label='Actual', s=80)

# Plotting the predictions
sns.scatterplot(x=X_test.flatten(), y=y_pred, color='red', label='Predicted', s=80)

plt.title('Decision Tree Regression on Iris Dataset')
plt.xlabel('Petal Width')
plt.ylabel('Iris Species')
plt.legend()

# Display the plot
plt.show()
```
